- Analisar taxa de entradas nulas e com 0's

```python
# Contar o n√∫mero de NAN e valores 0 nas tabelas
print("TAXA DE ENTRADAS NULAS")
print(economic.isna().sum()/economic.count())
print("-"*100)
print("TAXA DE ENTRADAS COM 0")
for column in economic.columns:
    print(column,":",round((economic[column]==0).sum()/economic.shape[0],2))
```

- Visualizar a distribui√ß√£o de cada coluna e checar se seguem uma distribui√ß√£o l√≥gica

```python
fig = economic.hist(bins=50, figsize=(10,7), xlabelsize=5, ylabelsize=5, grid=True)
[x.title.set_size(7) for x in fig.ravel()]
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
# Distribui√ß√£o das colunas normalizadas
dist_house_eco = economic[economic.columns[economic.columns.str.contains("TotalHouseholds")]]
pd.DataFrame(StandardScaler().fit_transform(
    dist_house_eco
        )).hist(bins=25)
plt.show()
```
