- Verificar outliers com Z-score, scatter plots e/ou interquartile

```python
def abs_z_score(column:pd.Series):
    return ((column-column.mean())/column.std()).abs()

demography[abs_z_score(demography["TotalPopulation"])>1]
```

```python
demography[abs_z_score(demography["MedianAgeInYears"])>3][["MedianAgeInYears"]].sort_values("MedianAgeInYears")
```

```python
def is_interquartile(column:pd.Series, percentiles=(25,75))->pd.Series:
    values = column.sort_values().dropna()
    q1,q3 = np.percentile(values, percentiles)
    return (column>q1)&(column<q3)
demography[is_interquartile(demography["SexRatio(males per 100 females)"],percentiles=(0.8,99.2))].sort_values("SexRatio(males per 100 females)")
```
